```python
#importign the networkx library as nx
import networkx as nx
```

```python
#creting the network graph
G = nx.Graph()
```

```python
#creting a dictionary with the person names and the index number.
#this is use to make the correct macthing edges and have some guidance with the index numbers 
labels = {0:"Andre",1:"Beverly",2:"Diane",3:"Carol",4:"Ed",5:"Fernando",6:"Garth",7:"Heather",8:"Ike",9:"Jane"}
```

```python
#now we start creating the edges between the nodes
G.add_edges_from([(0,1),(0,2),(0,3),(0,5),(1,2),(1,4),(1,6),(2,3),(2,4),(2,5),(2,6),(3,5),
                  (4,6),(5,6),(5,7),(6,7),(7,8),(8,9)])
```

```python
#checking the nodes
G.nodes()
```

```python
#checking the edges and confirm the right conection between the nodes
G.edges()
```

```python
#import the matplotlib library for drawing the graph 
import matplotlib.pyplot as plt
```

```python
#we draw the graph 
nx.draw(G)
plt.draw()
plt.show()
```

```python
#we draw the graph with the index labels
nx.draw(G,with_labels=True)
plt.draw()
plt.show()
```

```python
#in order to add the right labels to the nodes with we the relabel functions
G_new = nx.relabel_nodes(G, labels)
```

```python
#we draw the graph with the person labels
nx.draw(G_new,with_labels = True)
plt.draw()
plt.show()
```

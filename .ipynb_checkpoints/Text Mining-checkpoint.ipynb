### Subset and Cleaning Data 

Given that the dataset for the Yelp reviews is very large, we took into consideration computing memory limit and decided to work with a sample (n=1000) from the whole data. After taking a sample we created a function to clean the data and remove needless words for the training model.

```python
data_all = data[['text','stars']].dropna()
data_df = data_all.sample(n=10000)
len(data_df)
```

```python
#Stem word and stopword setting.
stemmer = stem.SnowballStemmer('english')
stopwords = set(stopwords.words('english'))
    
```

```python
#The cleaning function for removing Stem, punctuation and stop words.
def data_cleaning(data):
    data = data.lower()
    data = [message_word for message_word  in data.split() if message_word  not in stopwords]
    data = " ".join([stemmer.stem(message_word ) for message_word  in data])
    data = [message_word for message_word  in data.split() if message_word  not in string.punctuation]
    data = " ".join([stemmer.stem(message_word ) for message_word  in data])
    return data
```

## Data Preparation

We worked with Scikit-learn package when implementing our model predictor algorithms. We worked with four different methods: Naive Bayes, Random Forest Classifier, Stochastic Gradient Descent (SGD), and Logistic Regression.

### Spliting/Trainig Data

Before we started the modeling/training we put the data onto a matrix of TF-IDF. Later we continued splitting the data, with 80% being used for the training section and 20% for the testing.

```python
df_text = data_df['text'].apply(data_cleaning)

vectorizer = TfidfVectorizer(sublinear_tf=True, max_df=0.5,
                                       stop_words='english')
text = vectorizer.fit_transform(data_df['text'])

#Training/Splitting 80/20
X_train, X_test, y_train, y_test = train_test_split(text,data_df['stars'],test_size=0.2, shuffle=True,random_state=0,  stratify=data_df['stars'])
                                                    
                                                   
                                                
```

### Naive Bayes

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)

print("Accuracy: " + str(clf.score(X_test, y_test)))
```

### Random Forest Classifier

```python
# Set classifier 
RDC = RandomForestClassifier(n_estimators=100)

# Fit model
RDC.fit(X_train,y_train)

# Make predictions
prediction = RDC.predict(X_test)

# Score accuracy 
RDC_train_accuracy = round(RDC.score(X_train, y_train),4)
RDC_test_accuracy = round(RDC.score(X_test, y_test),4)

# Print accuracy 
print("Training Accuracy: " + str(RDC_train_accuracy))
print("Testing Accuracy: " + str(RDC_test_accuracy))
```

### Stochastic Gradient Descent (SGD) 

```python
# Set classifier 
SGD = SGDClassifier(max_iter=1000,tol=0.001)

# Make predictions
SGD.fit(X_train,y_train)

# Make predictions
prediction = SGD.predict(X_test)

# Score accuracy 
SGD_train_accuracy = round(SGD.score(X_train, y_train),4)
SGD_test_accuracy = round(SGD.score(X_test, y_test),4)

# Print accuracy 
print("Training Accuracy: " + str(SGD_train_accuracy))
print("Testing Accuracy: " + str(SGD_test_accuracy))
```

### Logistic Regression

## Sentiment Analysis - Labeling Data

For most of the text mining analysis remove , we worked with supervised data (labeled). On this project remove , we did not have labels for the sentiment analysis. We tried to solve this issue by labeling the data to see if we could get a better prediction model solution.

### Data Preparation

To prepare the data we created three labels (positive,negative, neutral) and tried to match them according the rating range (1-5). We split the labels as follows:
1. Rating >= 4 is positive 
2. Rating >= 2 and <4 is neutral
3. Rating <2 is negative

```python
#Creating the labeled classifications for the data.
conditions = [
    (data_df['stars'] >= 4),
    (data_df['stars']  >= 2) & (data_df['stars']  < 4),
    (data_df['stars']  < 2)]
choices = ['positive', 'neutral', 'negative']
data_df['sentiment'] = np.select(conditions, choices)
```

```python
#Creating and selecting labels positive and negative
data_new = data_df[(data_df['sentiment'] == 'positive') | (data_df['sentiment'] == 'negative')]
data_new.head()
```

```python
data_df.head()
```

### Splitting/Training Data

Before we started the modeling/training we put the data onto a matrix of TF-IDF. Later we continued splitting the data, with 80% being used for the training section and 20% for the testing.

```python
vectorizer = TfidfVectorizer(sublinear_tf=True, max_df=0.1,
                                       stop_words='english')
text = vectorizer.fit_transform(data_new['text'])

#Training/Splitting 80/20
X_train, X_test, y_train, y_test = train_test_split(text,data_new['sentiment'],
                                                    test_size=0.2, 
                                                    shuffle=True,
                                                    random_state=0,  
                                                    stratify=data_new['sentiment'])
    
```

### Logistic Regression

```python
log_model = LogisticRegression()
log_model = log_model.fit(X=X_train, y=y_train)
y_pred = log_model.predict(X_test)
print("Training Accuracy: " + str(log_model.score(X_train, y_train)))
print("Testing Accuracy: " + str(log_model.score(X_test, y_test)))
```

### Stochastic Gradient Descent (SGD) 


```python
# Set classifier 
SGD = SGDClassifier(max_iter=1000,tol=0.001)

# Make predictions
SGD.fit(X_train,y_train)

# Make predictions
prediction = SGD.predict(X_test)

# Score accuracy 
SGD_train_accuracy = round(SGD.score(X_train, y_train),4)
SGD_test_accuracy = round(SGD.score(X_test, y_test),4)

# Print accuracy 
print("Training Accuracy: " + str(SGD_train_accuracy))
print("Testing Accuracy: " + str(SGD_test_accuracy))
```

### Random Forest Classifier

```python
# Set classifier 
RDC = RandomForestClassifier(n_estimators=100)

# Fit model
RDC.fit(X_train,y_train)

# Make predictions
prediction = RDC.predict(X_test)

# Score accuracy 
RDC_train_accuracy = round(RDC.score(X_train, y_train),4)
RDC_test_accuracy = round(RDC.score(X_test, y_test),4)

# Print accuracy 
print("Training Accuracy: " + str(RDC_train_accuracy))
print("Testing Accuracy: " + str(RDC_test_accuracy))
```

### Naive Bayes

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)

print("Accuracy: " + str(clf.score(X_train, y_train)))
print("Accuracy: " + str(clf.score(X_test, y_test)))
```

## Analysis

### Challenges

One of the challenges was the size of the dataset. This dataset has 165,823 rows, which makes it very difficult to work with in any algorithm  due to the computing limit memory system. Another challenge was that the data was not labeled for sentiment analysis supervised algorithms.

## Conclusion

For this final project we implemented a sentiment analysis in an attempt to find out what comment  sentiments were and secondly to see if there was a relation between the three votes of useful, funny, cool and that particular comment. For example if the rating was 5 stars we expected that the comment was positive and vice versa.

Because our dataset was not labeled for sentiment analysis our first step on this project was to try to create some trained model based on the comment and the rating. After working with four different algorithm methods, we found out that our accuracy result was not strong enough. The accuracy result range was between 0.4 and 0.5, which we concluded would be very difficult to use for prediction. 

After that we then tried to separate and label the reviews using the data rating information. We did this by labeling 4-5 star ratings as positive sentiment, 2-3 stars as neutral, and 1 star as negative. This second attempt worked, but not as we expected because we again obtained low accuracy. We tried to find the best combination to obtain the maximum prediction modeling accuracy. The best model was obtained by labeling just the 4 and 5 star ratings as positive sentiment, and the 1 star as negative sentiment, which resulted in an accuracy result range ]between 0.92 - 0.95.

In the last training model for the most part we saw good testing results when trying to identify some comments with a rating of 1-4-5. When we use some of the neutral ratings of 2-3, the model was sometimes not good at deciding if it was a positive or negative sentiment.

We learned that text mining can be very useful for predicting ratings if we can get a well trained model. In this case we  worked with a not labeled dataset, which made it very challenging to work with a supervised learning system. It would be very interesting for further projects to try to work with neutral network which is good for unsupervised learning systems.
